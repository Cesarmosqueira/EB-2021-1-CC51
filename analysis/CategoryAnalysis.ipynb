```python
df = pd.read_csv('../data/cleaned_MXvideos_cc50_202101.csv')
```

```python
df[['likes', 'dislikes', 'category_description']]
likes = list(df['likes'])
dislikes = list(df['dislikes'])
cats = list(df['category_description'])
```

```python
ratio = []
for i in range(len(likes)):
    if dislikes[i] != 0:
        ratio.append(likes[i]/dislikes[i])
    else: ratio.append(dislikes[i])
```

```python
data = {'Category' :  cats, 'Ratio': ratio}
df = pd.DataFrame(data)
df = df.groupby(['Category'])[['Ratio']].mean()
```

```python
df.plot.bar(title="likes/dislikes",figsize=(15,10))
```

```python
df = pd.read_csv('../data/cleaned_MXvideos_cc50_202101.csv')
```

```python
df[['views', 'comment_count', 'category_description']]
views = list(df['views'])
cc = list(df['comment_count'])
cats = list(df['category_description'])
```

```python
ratio = []
for i in range(len(views)):
    if cc[i] != 0:
        ratio.append(views[i]/cc[i])
    else: ratio.append(cc[i])
```

```python
data = {'Category' :  cats, 'Ratio': ratio}
df = pd.DataFrame(data)
df = df.groupby(['Category'])[['Ratio']].mean()
```

```python
df.plot.bar(title="views/comment_count",figsize=(15,10))
```
